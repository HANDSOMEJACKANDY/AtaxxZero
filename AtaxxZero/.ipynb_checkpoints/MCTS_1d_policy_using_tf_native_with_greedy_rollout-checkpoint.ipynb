```python
def get_policy_dict_list():
    index=0
    policy_dict = {}
    policy_list = []
    for r in range(7):
        for c in range(7):
            for dr in range(-2, 3):
                for dc in range(-2, 3):
                    new_r = r + dr
                    new_c = c + dc
                    if (dr != 0 or dc != 0) and (new_r < 7 and new_r >= 0) and (new_c < 7 and new_c >= 0):
                        policy_dict[((r, c), (new_r, new_c))] = index
                        policy_list.append(((r, c), (new_r, new_c)))
                        index += 1
    return policy_dict, policy_list
policy_dict, policy_list = get_policy_dict_list()
```

```python
def get_rot_policy_dict():
    augment_dict = {}
    rot_m_0 = np.array([[0, -1], [1, 0]]) 
    center = np.array([3, 3])
    for is_flip in [False, True]:
        for rot_time in range(4):
            if not (is_flip == False and rot_time == 0):
                tmp_dict = {}
                # get rot matrix
                rot_m = np.eye(2)
                for i in range(rot_time):
                    rot_m = rot_m_0.dot(rot_m)

                for r in range(7):
                    for c in range(7):
                        for dr in range(-2, 3):
                            for dc in range(-2, 3):
                                start = np.array([r, c])
                                end = np.array([r+dr, c+dc])
                                if (dr != 0 or dc != 0) and \
                                    (start[0] < 7 and start[0] >= 0) and (start[1] < 7 and start[1] >= 0):
                                    new_start = (start - center)
                                    new_end = (end - center)
                                    if is_flip:
                                        new_start[1] = -new_start[1]
                                        new_end[1] = -new_end[1]
                                    new_start = rot_m.dot(new_start) + center
                                    new_end = rot_m.dot(new_end) + center
                                    tmp_dict[(tuple(start), tuple(end))] = (tuple(new_start), tuple(new_end))
                augment_dict[(is_flip, rot_time)] = tmp_dict
    return augment_dict
augment_dict = get_rot_policy_dict()
```

```python
def augment_policy(data, is_flip, rot_time):
    if is_flip == False and rot_time == 0:
        return data
    
    global policy_dict, policy_list, augment_dict
    out = np.zeros_like(data)
    for i in range(792):
        tmp_data = data[i]
        if tmp_data > 0:
            move = policy_list[i]
            move_after_rot = augment_dict[(is_flip, rot_time)][move]
            i_after_rot = policy_dict[move_after_rot]
            out[i_after_rot] = tmp_data
    return out
            
def augment_data(train_data):
    out = []
    # input is a list for follows
    feature_map = train_data[0]
    action_mask = train_data[1]
    frequency_map = train_data[2]
    
    # do 7 times augmentation
    for is_flip in [False, True]:
        for rot_time in range(4):
            # do feature map augmentation
            if is_flip:
                tmp_feature_map = np.fliplr(feature_map)
            tmp_feature_map = np.rot90(feature_map, k=rot_time, axes=(1, 2))
            # augment two policy related data
            tmp_action_mask = augment_policy(action_mask, is_flip, rot_time)
            tmp_frequency_map = augment_policy(frequency_map, is_flip, rot_time)
            # append them to out
            out.append([tmp_feature_map, tmp_action_mask, tmp_frequency_map])
    return out
```

```python
%%cython
import numpy as np
cimport numpy as np
import matplotlib.pyplot as plt
from random import choice

cdef class Ataxx:
    cdef public np.int8_t[:, :] data

    def __init__(self, board=None):
        if board is None:                  # if there is no initialization given
            self.data = np.zeros((7, 7), dtype=np.int8)   # then generate a board with starting init, and black(-1) takes first turn
            self.data[0, 0] = -1           
            self.data[6, 6] = -1
            self.data[0, 6] = 1
            self.data[6, 0] = 1
        else:
            self.data = board.copy()
            
    def reset(self):
        self.data = np.zeros((7, 7), dtype=np.int8)
        self.data[0, 0] = -1           
        self.data[6, 6] = -1
        self.data[0, 6] = 1
        self.data[6, 0] = 1
        
    def get_feature_map(self, turn):
        out = np.zeros((3, 9, 9))
        # define 1 edge
        out[0][:, :] = np.ones((9, 9), np.int8)
        out[0][1:-1, 1:-1] = np.zeros((7, 7), np.int8)
        # define friendly map
        out[1][1:-1, 1:-1] = (np.array(self.data) == turn).astype(np.int8)
        out[2][1:-1, 1:-1] = (np.array(self.data) == -turn).astype(np.int8)
        
        return out
    
    def plot(self, is_next_move=False, turn=None):                        # plot the board
        image = self.data.copy()
        if is_next_move:
            if turn not in [-1, 1]:
                raise ValueError("Turn must be -1 or 1, or Must input a turn for next moves")
            else:
                next_moves = self.get_moves(turn)
                if len(next_moves) == 0:
                    raise ValueError("Game is over already")
                next_pos = list(zip(*next_moves))[1]
                for pos in next_pos:
                    image[pos] = turn / 2
        plt.imshow(image, cmap='gray')
        plt.xticks(range(7), range(7))
        plt.yticks(range(7), range(7))
        plt.show()
                
    def is_valid(self, turn, pos):
        cdef int dr, dc, r = pos[0], c = pos[1], new_r, new_c
        if self.data[r, c] != 0:
            return False
        else:
            for dr in range(-2, 3):
                for dc in range(-2, 3):
                    new_r = r+dr
                    new_c = c+dc
                    if new_r >= 0 and new_c >= 0 and new_r < 7 and new_c < 7 and self.data[new_r, new_c] == turn:
                        return True
            return False 
        
    def get_moves(self, turn, return_corr_dict=False):
        cdef int r, c, dr, dc, new_r, new_c
        next_moves = []
        corr_dict = {}
        for r in range(7):
            for c in range(7):
                has_duplicate_move = False      # move within the radius of one of another friendly piece is called
                if self.is_valid(turn, (r, c)): # duplicate move
                    for dr in range(-2, 3):
                        for dc in range(-2, 3):
                            new_r = r+dr
                            new_c = c+dc
                            if new_r >= 0 and new_c >= 0 and new_r < 7 and new_c < 7 and self.data[new_r, new_c] == turn:
                                if abs(dr) <= 1 and abs(dc) <=1:
                                    if has_duplicate_move: 
                                        corr_dict[((new_r, new_c), (r, c))] = dup_move
                                    elif self.data[new_r, new_c] == turn:
                                        dup_move = ((new_r, new_c), (r, c))
                                        next_moves.append(dup_move) 
                                        has_duplicate_move = True
                                elif self.data[new_r, new_c] == turn:
                                    next_moves.append(((new_r, new_c), (r, c))) 
                                else:
                                    continue
        if return_corr_dict:
            return next_moves, corr_dict
        else:
            return next_moves
        
    
    def get_greedy_move(self, turn, moves=None):
        cdef int x0, y0, x1, y1, dr, dc, tmp_score, best_score = -50
        # get all possible moves if not provided
        if moves is None:
            moves, corr_dict = self.get_moves(turn, return_corr_dict=True)
            for item in corr_dict:
                moves.append(item)
        
        if len(moves) == 0:
            raise ValueError('No Possible Moves')
        
        best_moves = []
        # calculate greedy move
        for (x0, y0), (x1, y1) in moves:
            tmp_score = 0
            if abs(x0-x1) <= 1 and abs(y0-y1) <= 1:
                tmp_score += 1
            for dr in range(-1, 2):
                for dc in range(-1, 2):
                    try:
                        if x1+dr >= 0 and y1+dc >= 0:
                            tmp_score += self.data[x1+dr, y1+dc] == -turn
                    except:
                        pass
            if tmp_score > best_score:
                best_moves = [((x0, y0), (x1, y1))]
                best_score = tmp_score
            elif tmp_score == best_score:
                best_moves.append(((x0, y0), (x1, y1)))
        return choice(best_moves)
        
    def move_to(self, turn, pos0, pos1):
        cdef int dr, dc, x0 = pos0[0], y0 = pos0[1], x1 = pos1[0], y1 = pos1[1]
        
        if not self.is_valid(turn, pos1):
            raise ValueError("This move: " + str((pos0, pos1)) + " of turn: " + str(turn) + " is invalid") 
        elif self.data[x0, y0] != turn:
            raise ValueError("The starting position is not your piece")
        else:
            self.data[x1, y1] = turn
            if abs(x0 - x1) > 1 or abs(y0 - y1) > 1:   # jump move
                self.data[x0, y0] = 0

            for dr in range(-1, 2):                  # infection mode!!!!
                for dc in range(-1, 2):
                    if x1+dr >= 0 and y1+dc >= 0 and x1+dr < 7 and y1+dc < 7:
                        if self.data[x1+dr, y1+dc] == -turn:  # convert any piece of the opponent to 'turn'
                            self.data[x1+dr, y1+dc] = turn
    
    def evaluate(self, turn, this_turn, max_score=1, min_score=0.001):
        cdef int r, c, turn_no=0, op_no=0
        for r in range(7):
            for c in range(7):
                if self.data[r, c] == turn:
                    turn_no += 1
                elif self.data[r, c] == -turn:
                    op_no += 1
        if len(self.get_moves(this_turn)) == 0:# if one of them can no longer move, count and end
            if turn_no > op_no:
                return max_score
            else:
                return -max_score
        else:
            value = turn_no - op_no
        return value * min_score
```

```python
@nb.jit(nopython=True)
def for_none(c, parent_n_visit, p):
    #return c * sqrt(2 * log(parent_n_visit+1)) + np.random.random()*1e-4 + p
    return c * (p + np.random.random()*1e-6) + 0.5 #+ 0.85 * sqrt(2 * log(parent_n_visit+1))

@nb.jit(nopython=True)
def for_node(c, parent_n_visit, n_visit, n_win, p):
    #return (n_win) / (n_visit+1e-5) + c * sqrt(2 * log(parent_n_visit+1) / (n_visit+1)) + np.random.random()*1e-4 + p
    return (n_win) / (n_visit+1e-5) + c * (p + np.random.random()*1e-6) / (n_visit+1) #+ 0.85 * sqrt(2 * log(parent_n_visit+1) / (n_visit+1))
        
class TreeNode():
    def __init__(self, parent, p=0.0):
        self._parent = parent
        self._children = {} # a dictionary of action:node
        self._n_visit = 0.0
        self._n_win = 0.0 # the win here is from the parent perspective
        self._p = p
        self._action_mask = None
        
    def __str__(self):
        out = "_n_visit: {}, _n_win: {}, _p: {}, _children: \n{}".format(\
                self._n_visit, self._n_win, self._p, self._children)
        return out
    
    def access_children(self, move):
        out = self._children[move]
        if type(out) is tuple:
            out = self._children[out]
        return out
    
    def children_generator(self):
        for move in self._children:
            yield (move, self.access_children(move))
    
    def update_all(self, is_win):
        node = self
        while not node is None:
            is_win = not is_win # intended
            node._n_win += is_win
            node._n_visit += 1
            node = node._parent
            
    @staticmethod
    def get_search_value(parent, node, c):
        # give the chance to root (simulate greedy attributes)
        if type(node) is tuple:
            node = parent._children[node]
        # return values 
        if type(node) is np.float32:
            # log and sqrt in math module is super fast
            value = for_none(c, parent._n_visit, node)
        else:
            value = for_node(c, parent._n_visit, node._n_visit, node._n_win, node._p)
        #print(value)
        return value
        
    @staticmethod
    def get_frequency_value(node):
        try:
            return node._n_visit
        except:
            return 0
    
    def select(self, c):
        best_node = [1, 2]
        best_node[0], best_node[1] = max(self._children.items(), key=lambda node: self.get_search_value(self, node[1], c))
        if type(best_node[1]) is tuple:
            best_node[0] = best_node[1] # update to the root method
            best_node[1] = self._children[best_node[1]] # update to the root node (or _p)
        if type(best_node[1]) is np.float32:
            # update _p attribute
            new_node = TreeNode(self, best_node[1])
    
            assert not type(self._children[best_node[0]]) is tuple
        
            # give this node to the root of this move
            self._children[best_node[0]] = new_node
            return (best_node[0], new_node)
        else:
            return best_node
        
    def get_action_mask(self):
        # only generate the action mask once
        if not self._action_mask is None:
            return self._action_mask
        else:
            global policy_dict
            out = np.zeros(len(policy_dict), dtype=np.int8)
            # record all reasonable action
            for move in self._children:
                out[policy_dict[move]] = 1
            return out
    
    def get_action_frequency_map(self):
        global policy_dict
        out = np.zeros(len(policy_dict))
        # record all the n_visit of each node
        nodes = self.children_generator()
        for node in nodes:
            if not type(node[1]) is np.float32:
                out[policy_dict[node[0]]] = node[1]._n_visit
        # normalize the array
        out /= out.sum()
        return out
```

```python
class PolicyNetwork():
    def __init__(self, lr=None, is_load_model=False):
        self._sess = tf.Session(config=tf.ConfigProto(intra_op_parallelism_threads=24))
        K.set_session(self._sess)
        
        if is_load_model:
            self._model = load_model('MCTS_POLICY_MODEL/train_model.h5')
            self._target_model = load_model('MCTS_POLICY_MODEL/target_model.h5')
            print("successfully loaded two models")
        else:
            assert not lr is None
            self._lr = lr
            self._model = self.create_model()
            self._target_model = self.create_model()
            print("new models generated")
            
        # synchronize both models
        self.update_target_model()
        # setting for tf forward propagation
        init = tf.global_variables_initializer()
        self._sess.run(init)
        # print the model structure
        print(self._model.summary())
        
    def update_learning_rate(self, lr):
        if self._lr != lr:
            self._model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=lr, decay=1e-7))
            self._lr = lr
            print("learning rate updated from {} to {}".format(self._lr, lr))
        
    def create_model(self):
        assert K.backend() == 'tensorflow'

        with tf.device('cpu'):
            board_input = Input((3, 9, 9))
            mask_input = Input((792, ))
            
            x = Conv2D(32, (3, 3), activation='relu', padding='valid')(board_input)
            x = BatchNormalization(axis=1)(x)
            x = Dropout(0.25)(x)
            x = Conv2D(64, (3, 3), activation='relu', padding='valid')(x)
            x = BatchNormalization(axis=1)(x)
            x = Dropout(0.25)(x)
            x = Conv2D(32, (3, 3), activation='relu', padding='valid')(x)
            x = BatchNormalization(axis=1)(x)
            x = Flatten()(x)
            x = Dense(792, activation='softplus')(x)
            x = multiply(inputs=[x, mask_input])     # this mask will mask any illegal move
            action_output = Activation('softmax')(x)

            model = Model(input=[board_input, mask_input],output=[action_output])
            model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=self._lr, decay=1e-7))
        
        return model

    def update_target_model(self):
        model_weights = self._model.get_weights()
        self._target_model.set_weights(model_weights)
        
    def predict(self, feature_map, action_mask, is_target=False):
        if not is_target:
            model = self._model
        else:
            model = self._target_model
        
        return self._sess.run(model.output, feed_dict={model.inputs[0]: feature_map.reshape(-1, 3, 9, 9), \
                                    model.inputs[1]: action_mask.reshape(-1, 792), K.learning_phase(): 0})[0]
    
    def save(self):
        self._model.save('MCTS_POLICY_MODEL/train_model.h5')
        self._target_model.save('MCTS_POLICY_MODEL/target_model.h5')
```

```python
%%cython
import numpy as np
cimport numpy as np
def get_argmax(np.ndarray[np.int8_t] action_mask, np.ndarray[np.float32_t] p_array):
    sort_p_array = action_mask * (0.75 * p_array + 0.25 * np.random.dirichlet(0.3*np.ones(p_array.shape[0])))
    # sort_p_array = action_mask * (0.8 * p_array + 0.2 * np.random.random(p_array.shape[0]))
    return sort_p_array.argmax()
```

```python
class Relay():
    def __init__(self, life_span=6):
        self._relay_dict = {}
        self._counter = 0
        self._life_span = life_span
        
    def append(self, train_data):
        train_data = list((zip(*train_data)))
        # stack the small batch
        for i in range(3):
            train_data[i] = np.stack(train_data[i], axis=0)
        # append new data
        self._relay_dict[self._counter] = train_data
        self._counter += 1
        
        # remove too old data
        remove_index = -1
        for index in self._relay_dict:
            if self._counter - index > self._life_span:
                remove_index = index
        if remove_index != -1:
            del self._relay_dict[remove_index]
                
    def get(self):
        out = [[], [], []]
        for _, data in self._relay_dict.items():
            # grab all data still living
            out[0].append(data[0])
            out[1].append(data[1])
            out[2].append(data[2])
            
        for i in range(3):
            out[i] = np.concatenate(out[i], axis=0)
        
        return out
```

```python
class MCTS():
    def __init__(self, c=1, slow_step=20, lr=1e-3, is_load_model=False):
        # slow_step means how many step we use to do typical mcts, after that we do fast play
        self._root = TreeNode(None) # this one will move in self play mode
        self._root_store = self._root # this is a backup for reset
        self._c = c
        self._game = Ataxx()
        self._turn = -1
        self._slow_step = slow_step
        # generate model
        if self._c > 0:
            self._policy = PolicyNetwork(lr, is_load_model)
        self._train_relay = Relay(life_span=12)
        
    def reset(self, left_space=45):
        self._game.reset()
        self._turn = -1
        self._root = self._root_store
        
        if left_space < 45:
            steps = 0
            is_terminal = False
            result = 45
            while not is_terminal and result > left_space:
                if np.random.random() < 0.2: # 80 percent using greedy move
                    best_move = choice(self._game.get_moves(self._turn))
                    
                else:
                    best_move = self._game.get_greedy_move(self._turn)
                    
                self.make_a_move(best_move)
                is_terminal = abs(self._game.evaluate(1, self._turn)) == 1
                result = (np.array(self._game.data) == 0).sum()
                steps += 1
            if is_terminal:
                print("reset failure, do reset again")
                self.reset(left_space)
        print("reset successful")
        
        
    def expand(self, node, game, turn):
        new_moves, corr_dict = game.get_moves(turn, return_corr_dict=True)
        for move in new_moves:
            node._children[move] = None # do not create a node until it is explored
        for move in corr_dict:
            node._children[move] = corr_dict[move]

    def plot_move_visit_freq(self):
        nodes = sorted(self._root._children.items(), key=lambda node: self._root.get_frequency_value(node[1]), reverse=True)
        # nodes = self._root.children_generator()
        for node in nodes:
            try:
                print(node[0], node[1]._n_win, node[1]._n_visit)
            except:
                print(node[0], 0, 0)
                pass
            
    def get_p_for_node(self, node, game, turn):
        # check if is leaf node
        if node._children == {}:
            return
        
        # get action_mask
        if self._c > 0:
            global policy_dict
            action_mask = node.get_action_mask()
            # get feature map
            feature_map = game.get_feature_map(turn)
            # give p for children
            p_array = self._policy.predict(feature_map, action_mask, is_target=True)

            # normalize p_array to mean = 1: this will not directly increase the mean to 1
            # it will grow as the network learn and stop to spread the prob everywhere
            p_array *= action_mask.sum()
        else:
            p_array = np.zeros(792, dtype=np.float32)
    
        for move in node._children:
            if not type(node._children[move]) is tuple:
                node._children[move] = p_array[policy_dict[move]]
    
    def get_fast_move(self, feature_map, action_mask):
        global policy_list
           
        # implement prediction
        p_array = self._policy.predict(feature_map, action_mask, is_target=True)
        max_index = get_argmax(action_mask, p_array)

        return policy_list[max_index]
        
    
    def fast_play(self, node, game, turn, is_simple=True):
        global policy_dict
        n_step = 0
        # generate current actions
        tmp_turn = turn
        if not is_simple and self._c > 0:
            action_mask = np.zeros(len(policy_dict), dtype=np.int8)
        moves, corr_dict = game.get_moves(tmp_turn, True)
        while len(moves) > 0:    
            # get a policy fast move and make the move
            if not is_simple and self._c > 0:
                # generate current action mask
                for move in moves:
                    action_mask[policy_dict[move]] = 1
                for move in corr_dict:
                    action_mask[policy_dict[move]] = 1
                # generate current feature map
                feature_map = game.get_feature_map(tmp_turn)
                move = self.get_fast_move(feature_map, action_mask)
            else:
                for item in corr_dict:
                    moves.append(item)
                if np.random.random() > 0.1:
                    move = choice(moves)
                else:
                    move = game.get_greedy_move(tmp_turn, moves)
            
            # debug purpose
            if n_step > 300:
                game.plot()

            # move = choice(moves)
            game.move_to(tmp_turn, move[0], move[1])
            # refresh the game
            tmp_turn = -tmp_turn
            # get new moves before getting action mask
            if not is_simple:
                action_mask[:] = 0
            moves, corr_dict = game.get_moves(tmp_turn, True)
            n_step += 1
            
        # as the game is over, see if turn has won or not
        turn_no = 0
        op_no = 0
        for r in range(7):
            for c in range(7):
                if game.data[r, c] == turn:
                    turn_no += 1
                elif game.data[r, c] == -turn:
                    op_no += 1
        is_win = turn_no > op_no
        return is_win, n_step

    
    def rollout(self, rollout_times=100, t_lim=np.nan):
        start = time.time()
        for i in range(rollout_times):
            tmp_node = self._root
            tmp_game = Ataxx(self._game.data)
            tmp_turn = self._turn
            # step counter init
            step = 0
            while True:
                if step < self._slow_step:
                    # expand tree
                    # init a node init!!!!!!!!
                    if tmp_node._children == {}:
                        # expand only moves
                        self.expand(tmp_node, tmp_game, tmp_turn)
                        # update p for each child if there are
                        self.get_p_for_node(tmp_node, tmp_game, tmp_turn)

                    # check if is leaf node, if so, update the whole tree
                    if tmp_node._children == {}:
                        turn_no = 0
                        op_no = 0
                        for r in range(7):
                            for c in range(7):
                                if tmp_game.data[r, c] == tmp_turn:
                                    turn_no += 1
                                elif tmp_game.data[r, c] == -tmp_turn:
                                    op_no += 1
                        is_win = turn_no > op_no
                        tmp_node.update_all(is_win) # ths is_win here is reversed for parent update
                        break
                    else:
                        # select a child and continue exploration
                        next_move, next_node = tmp_node.select(self._c)
                            
                        # move to next move and next node
                        tmp_game.move_to(tmp_turn, next_move[0], next_move[1])    
                        tmp_node = next_node
                        tmp_turn = -tmp_turn
                else:
                    is_win, d_step = self.fast_play(tmp_node, tmp_game, tmp_turn, is_simple=True)
                    step += d_step
                    tmp_node.update_all(is_win)
                    break
                # update steps                                    
                step += 1
            # print step no
            # print("total step: ", step)
            if time.time() - start > t_lim:
                print(i)
                break
                      
    def get_next_move(self, rollout_times=100, t_lim=np.nan):
        # do mcts
        self.rollout(rollout_times, t_lim)
        # return the best move
        best_node = max(self._root._children.items(), key=lambda node: self._root.get_frequency_value(node[1]))

        return best_node[0]
    
    def make_a_move(self, next_move):
        # move the root to next_move's root
        if self._root._children == {}:
            self._root = TreeNode(None)
        elif type(self._root.access_children(next_move)) is np.float32: # the root may not be neccessarily expanded
            out = self._root._children[next_move]
            if type(out) is tuple:
                self._root._children[out] = TreeNode(None)
            else:
                self._root._children[next_move] = TreeNode(None)
            self._root = self._root.access_children(next_move)
        else:
            self._root = self._root.access_children(next_move)
        
        self._root._parent = None
        
        # update the game board
        self._game.move_to(self._turn, next_move[0], next_move[1])
        self._turn = -self._turn
        
    def testing_against_pure_mcts(self, rounds=5, left_space=None, rival_rollout_times=400, self_rollout_times=400, verbose=True):
        print("####               ####")
        print("#### start testing ####")
        # recorder of game result
        n_win = 0
        # initialize a player with no policy
        tmp_player = MCTS(c=0, slow_step=20)
        for r in range(rounds):
            tmp_round_s = time.time()
            print("round:", r+1)
            # randomly init the game board if no left_space specified
            if left_space is None:
                left_space = np.random.randint(30, 46)
            self.reset(left_space=left_space)
            print("this game start with {} space left".format(left_space))
            # synchronize both games
            tmp_player._game = Ataxx(self._game.data)
            tmp_player._turn = self._turn
            # set up start turns
            tmp_player_turn = choice([-1, 1])
            if verbose:
                print("self takes turn: ", -tmp_player_turn)
            # start the game
            steps = 0
            while abs(self._game.evaluate(1, self._turn)) != 1:
                # plot the game board
                if verbose:
                    self._game.plot()
                    tmp_s = time.time()
                if self._turn == tmp_player_turn:
                    best_move = tmp_player.get_next_move(rival_rollout_times)
                    if verbose:
                        print("rival turn")
                        print(tmp_player.plot_move_visit_freq())
                else:
                    best_move = self.get_next_move(self_rollout_times)
                    if verbose:
                        print("self turn")
                        print(self.plot_move_visit_freq())
                if verbose:
                    print("this move takes time(s): ", time.time()-tmp_s)
                    print("chosen move is ", best_move)

                # synchronize steps and boards
                self.make_a_move(best_move)
                tmp_player.make_a_move(best_move)
                # update steps
                steps += 1
                if steps > 250:
                    print("steps over 250, game skip")
                    break
            if steps <= 250:
                is_self_win = self._game.evaluate(-tmp_player_turn, self._turn) == 1
                if is_self_win:
                    n_win += 1
                print("this round has steps: {}, time taken: {}, self wins? {}".format(steps, time.time()-tmp_round_s, is_self_win))
            else:
                n_win += 0.5
        return n_win / float(rounds)

                
    def self_play(self, rollout_times=100, t_lim=np.nan, verbose=True):
        train_data = []
        steps = 0
        print("start new self play")
        while abs(self._game.evaluate(1, self._turn)) != 1 and steps < 250:
            print(self._turn, "'s turn")
            tmp_s = time.time()
            best_move = self.get_next_move(rollout_times, t_lim=np.nan)
            print("one move takes time(s): ", time.time()-tmp_s)
            # record π data
            tmp_data = [self._game.get_feature_map(self._turn), \
                        self._root.get_action_mask(), \
                        self._root.get_action_frequency_map()]
            tmp_data = augment_data(tmp_data)
            train_data.extend(tmp_data)
            # plot the game board if verbose
            if verbose:
                if np.random.random() < 0.01:
                    self._game.plot()
                    self.plot_move_visit_freq()
            # maker the move and move on
            self.make_a_move(best_move)
            steps += 1
        print("this self play has {} steps".format(steps))
        return train_data
            
    def reinforcement_learning(self, episode=1000, rollout_times=100, left_space_max=46):        
        for epi in range(episode):
            print("episode {} now start".format(epi))
            # randomly skip a few steps if left space not specified
            left_space = np.random.randint(left_space_max/2, left_space_max)
            self.reset(left_space=left_space)
            print("left space is {}".format(left_space))
            # start self_play and get train_data
            train_data = self.self_play(rollout_times=rollout_times)
            # store it in relay
            self._train_relay.append(train_data)
            # do training every 5 epi and update target model and reset tree
            if epi%3 == 0:
                train_data = self._train_relay.get()
                print("start training, training data no. {}".format(train_data[0].shape[0]))
                # do training
                self._policy._model.fit(x=[train_data[0], train_data[1]], y=[train_data[2]], verbose=1, \
                                            batch_size=256, epochs=512, shuffle=True)
                self._policy.update_target_model()
                self._root = TreeNode(None)
                self._root_store = self._root
                if epi%6 == 0:
                    win_rate = self.testing_against_pure_mcts(10, 10, 100, 100, False)
                    print("testing win rate is: ", win_rate)
            print("saving files")
            self._policy.save()
            print("episode {} finished".format(epi))
```
